### Missing Values

Missing value analysis is performed to identify variables with incomplete data and measure the extent of missing observations in the dataset. Detecting missing values is an important step in assessing data quality, as incomplete data can affect the accuracy of statistical analysis, data visualization, and machine learning models if left untreated.

```python
# @title
df.isnull().sum()
```

```python
# @title
missing_values = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage (%)": (df.isnull().sum() / len(df) * 100).round(2)
})

missing_values
```
