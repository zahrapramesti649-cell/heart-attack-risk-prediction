### Data Cleaning & Preprocessing

### Objective

This notebook aims to prepare a clean and structured dataset for exploratory data analysis (EDA) and machine learning modeling. The focus is on handling data quality issues such as missing values, ensuring appropriate data formats, and validating the overall dataset structure.

The output of this stage is a cleaned dataset that is consistent, reliable, and ready for further analysis and model development.

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/heart_attack_dataset.csv", sep=";")
df.head()
```

### Missing Values Overview

Before proceeding with data cleaning, a re-evaluation of missing values is conducted to better understand their distribution and extent across variables. This step is essential for determining the most appropriate handling strategy, such as imputation or removal, to ensure data quality for subsequent analysis and modeling.

```python
missing_values = df.isnull().sum()
missing_values
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
cat_cols = df.select_dtypes(include=['object']).columns

df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
```

```python
df.isnull().sum()
```

### Data Cleaning Result

After handling missing values, all variables in the dataset are now complete with no missing entries. The imputation strategy applied includes using the median for numerical variables and the mode for categorical variables, as these approaches are robust to data distribution and suitable for preserving the overall characteristics of the dataset.

With these adjustments, the dataset is now clean, consistent, and ready for Exploratory Data Analysis (EDA).
