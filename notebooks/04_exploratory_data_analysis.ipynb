### Objective

Exploratory Data Analysis (EDA) aims to uncover patterns, relationships, and distributions within the dataset. This stage focuses on exploring key variables related to heart attack risk in order to generate initial insights that support both business understanding and predictive modeling.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/heart_attack_dataset.csv", sep=";")
```

```python
df['heart_attack_risk'].value_counts()
```

```python
sns.countplot(x='heart_attack_risk', data=df)
plt.title("Distribution of Heart Attack Risk")
plt.show()
```

### Target Distribution

The target distribution represents the proportion of patients with heart attack risk (1) and without risk (0). Understanding this distribution is crucial to identify whether the dataset is balanced or imbalanced, as class imbalance can significantly affect the performance and bias of machine learning models.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='heart_attack_risk', y='age', data=df)
plt.title("Age vs Heart Attack Risk")
plt.show()
```

### Age vs Heart Attack Risk

Patients at risk of heart attack tend to be older compared to those without risk, indicating that age may be a contributing factor to heart attack likelihood.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='heart_attack_risk', y='cholesterol', data=df)
plt.title("Cholesterol vs Risk")
plt.show()
```

### Cholesterol vs Risk

There is a tendency for patients at risk of a heart attack to have higher cholesterol levels.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='heart_attack_risk', y='bmi', data=df)
plt.title("BMI vs Risk")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=['int64','float64']).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.countplot(x='smoking_status', hue='heart_attack_risk', data=df)
plt.xticks(rotation=45)
plt.show()
```
